```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import cv2# opencv


```

```python
img=cv2.imread(car_img_path)
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
img2=cv2.cvtColor(img,cv2. COLOR_BGR2RGB)
```

```python
img2.shape
```

```python
plt.imshow(img2)
```

```python
img3=cv2.cvtColor(img,cv2. COLOR_BGR2GRAY)
```

```python
img3.shape
```

```python
plt.imshow(img3,cmap='gray')
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(img)
```

```python
img3.shape
```

```python
row,col=img3.shape
```

```python
car=img3.reshape(1,row,col,1)
```

```python
car.shape
```

```python
conv_layer=[tf.keras.layers.Conv2D(
    filters=1,
    kernel_size=(3,3),
    strides=(1,1),
    input_shape=car.shape[1:]

)]
```

```python
conv_model=tf.keras.Sequential(conv_layer)
```

```python
conv_model.summary()
```

```python
output = conv_model.predict(car)
```

```python
output.shape
```

```python
row,col=output.shape[1:-1]
```

```python
reshaped_output=output.reshape(row,col)
```

```python
plt.imshow(reshaped_output,cmap='gray')
```
